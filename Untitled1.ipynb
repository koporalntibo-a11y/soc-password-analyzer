```python
import tkinter as tk
from tkinter import messagebox
import sqlite3
import re
import random
import string
import matplotlib.pyplot as plt
import smtplib

# ===================== DATABASE SETUP =====================
conn = sqlite3.connect("passwords.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS password_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    password TEXT,
    score INTEGER,
    risk_level TEXT
)
""")

conn.commit()

# ===================== BREACH DATABASE (SIMULATED) =====================
breached_passwords = ["123456", "password", "admin", "qwerty", "letmein"]

# ===================== PASSWORD ANALYZER =====================
def analyze_password(password):
    score = 5
    issues = []

    if len(password) < 8:
        score -= 1
        issues.append("Too short")

    if not re.search(r"[A-Z]", password):
        score -= 1
        issues.append("No uppercase")

    if not re.search(r"[a-z]", password):
        score -= 1
        issues.append("No lowercase")

    if not re.search(r"[0-9]", password):
        score -= 1
        issues.append("No numbers")

    if not re.search(r"[!@#$%^&*()]", password):
        score -= 1
        issues.append("No special characters")

    if password.lower() in breached_passwords:
        score = 0
        issues.append("COMMON PASSWORD DETECTED")

    if score >= 4:
        level = "LOW RISK"
    elif score >= 2:
        level = "MEDIUM RISK"
    else:
        level = "HIGH RISK"

    return score, level, issues

# ===================== BREACH CHECK =====================
def check_breach(password):
    return password.lower() in breached_passwords

# ===================== SAVE TO DATABASE =====================
def save_to_db(password, score, level):
    cursor.execute(
        "INSERT INTO password_logs (password, score, risk_level) VALUES (?, ?, ?)",
        (password, score, level)
    )
    conn.commit()

# ===================== NORMAL PASSWORD GENERATOR =====================
def generate_password():
    words = ["sun", "blue", "river", "code", "lion", "cloud", "star", "tech"]
    pwd = random.choice(words) + str(random.randint(10, 99))
    entry.delete(0, tk.END)
    entry.insert(0, pwd)

# ===================== EMAIL ALERT (SIMULATION) =====================
def send_email_alert(password, level):
    if level == "HIGH RISK":
        print("ALERT: High-risk password detected (email simulation)")

# ===================== ANALYTICS DASHBOARD =====================
def show_analytics():
    cursor.execute("SELECT risk_level FROM password_logs")
    data = cursor.fetchall()

    low = sum(1 for d in data if d[0] == "LOW RISK")
    medium = sum(1 for d in data if d[0] == "MEDIUM RISK")
    high = sum(1 for d in data if d[0] == "HIGH RISK")

    plt.bar(["Low", "Medium", "High"], [low, medium, high])
    plt.title("SOC Password Risk Distribution")
    plt.show()

# ===================== MAIN ANALYSIS =====================
def analyze():
    password = entry.get()

    if not password:
        messagebox.showerror("Error", "Enter a password")
        return

    score, level, issues = analyze_password(password)
    breached = check_breach(password)

    if breached:
        level = " CRITICAL BREACH DETECTED"

    save_to_db(password, score, level)
    send_email_alert(password, level)

    result = f"""
Password: {password}
Score: {score}/5
Risk Level: {level}
Issues: {', '.join(issues)}
Breached: {"YES" if breached else "NO"}
"""

    messagebox.showinfo("SOC ANALYSIS RESULT", result)

# ===================== HACKER STYLE GUI =====================
app = tk.Tk()
app.title("SOC Security Terminal")
app.geometry("600x420")
app.configure(bg="#0b0f14")

# HEADER
header = tk.Label(
    app,
    text="🛡 SOC PASSWORD ANALYZER TERMINAL",
    font=("Consolas", 16, "bold"),
    fg="#00ff9f",
    bg="#0b0f14"
)
header.pack(pady=15)

# FRAME
frame = tk.Frame(app, bg="#111827", bd=2, relief="ridge")
frame.pack(pady=10, padx=20, fill="both", expand=True)

# LABEL
tk.Label(
    frame,
    text="ENTER PASSWORD:",
    font=("Consolas", 12),
    fg="#00ff9f",
    bg="#111827"
).pack(pady=10)

# ENTRY
entry = tk.Entry(
    frame,
    width=35,
    font=("Consolas", 14),
    bg="#0b0f14",
    fg="#00ff9f",
    insertbackground="#00ff9f",
    relief="flat"
)
entry.pack(pady=5)

# BUTTON STYLE
def make_button(text, command):
    return tk.Button(
        frame,
        text=text,
        command=command,
        font=("Consolas", 11, "bold"),
        bg="#00ff9f",
        fg="#0b0f14",
        activebackground="#00cc7a",
        activeforeground="black",
        relief="flat",
        width=25,
        pady=5
    )

# BUTTONS
make_button(" ANALYZE PASSWORD", analyze).pack(pady=8)
make_button(" GENERATE PASSWORD", generate_password).pack(pady=8)
make_button(" SHOW ANALYTICS", show_analytics).pack(pady=8)

# FOOTER
tk.Label(
    app,
    text="SOC Simulation System v1.0 | Developed by SM NTIBO",
    font=("Consolas", 10),
    fg="#666",
    bg="#0b0f14"
).pack(side="bottom", pady=10)
app.mainloop()
```
